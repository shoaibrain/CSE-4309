```python
tran_data = pd.read_csv("transactions_train.csv")
customer_data = pd.read_csv("customers.csv")
article_data = pd.read_csv("articles.csv")
```

```python
merge_phase_1 = pd.merge(
    tran_data, #left
    customer_data, #right
    how="inner",
    on="customer_id",
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None, #TODO: validate many_to_one ratio
)
```

```python
merge_phase_2 = pd.merge(
    merge_phase_1, #left
    article_data, #right
    how="inner",
    on="article_id",
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None, #TODO: validate many_to_one ratio
)
```
